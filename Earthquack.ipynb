```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
# load the dataset
df = pd.read_csv('Untitled Folder/earthquake_alert_balanced_dataset.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
x = df[df.columns[:-1]]
```

```python
y = df[df.columns[-1]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
knn_model = KNeighborsClassifier()
nb_model = GaussianNB()
svm_model = SVC()
```

```python
knn_model.fit(x_train,y_train)
nb_model.fit(x_train,y_train)
svm_model.fit(x_train,y_train)
```

```python
knn_pred = knn_model.predict(x_test)
nb_pred = nb_model.predict(x_test)
svm_pred = svm_model.predict(x_test)
```

```python
print(classification_report(y_test,knn_pred))
```

```python
nb_x_train,nb_x_test,nb_y_train,nb_y_test = train_test_split(x,y,test_size=0.2,random_state=100)
nb_model.fit(nb_x_train,nb_y_train)
nbx_pred = nb_model.predict(nb_x_test)
print(classification_report(y_test,nb_pred))
```

```python
svm_x_train,svm_x_test,svm_y_train,svm_y_test = train_test_split(x,y,test_size = 0.2 , random_state=83)
svm_model.fit(svm_x_train,svm_y_train)
svm_pred = svm_model.predict(svm_x_test)
print(classification_report(svm_y_test,svm_pred))
```
